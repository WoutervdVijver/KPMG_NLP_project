```python
# Accessing the txt file containing the sample nl text
# Importing file operations Python library
import os

file = open("nl_sample.txt","rt")
raw_text = file.read()
file.close()
```

```python
print(raw_text)
```

```python
# Removing white space and putting the text in lower case
raw_text = raw_text.lower()
raw_text = raw_text.replace('\n','')
print(raw_text)
```

```python
raw_text = raw_text.replace('\x0c','')
print(raw_text)
```

```python
# Load the dutch model (small version) of Spacy

import spacy
nlp = spacy.load("nl_core_news_sm")
```

```python
# Store the tokens in doc

doc = nlp(raw_text)
```

### Tokenizing text that has been rid of white space and put in lower case

```python
for token in doc:
    print(token)
```

```python
len(nlp(raw_text)) 
```

### Part-of-speech tagging

```python
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)
```

** VISUALISING POS TAGS**

```python
import spacy
from spacy import displacy

nlp = spacy.load("nl_core_news_sm")
doc = nlp(raw_text)
displacy.render(doc, style="dep")
```

### Morphology

```python
nlp = spacy.load('nl_core_news_sm')
print("Pipeline:", nlp.pipe_names)

doc = nlp(raw_text)
token = doc[0]  # 'I'
print(token.morph)  # 'Case=Nom|Number=Sing|Person=1|PronType=Prs'
print(token.morph.get("PronType"))  # ['Prs']
```

### Unique characters

```python
# Capture the unique characters
unique_characters = set(raw_text)
print(unique_characters)
```

```python
# Putting them in a string
unique_characters_in_string = ''.join(set(raw_text))
print(unique_characters_in_string)
```
